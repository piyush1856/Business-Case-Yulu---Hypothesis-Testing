4. **Hypothesis Testing**
- **Working Day Effect on Rental Demand (2-Sample T-Test)**:

  - `Hypotheses`:
    - Null Hypothesis (H₀): There is no significant difference in bike rentals between working days and non-working days.
    - Alternate Hypothesis (H₁): There is a significant difference.
  - `Test`: Perform a two-sample independent T-test to compare the average rentals on working days versus non-working days. Based on the p-value, accept or reject the null hypothesis.
- **Weather Effect on Rental Demand (One-Way ANOVA)**:

  - `Hypotheses`:
    - Null Hypothesis (H₀): Bike rentals are the same across different weather conditions.
    - Alternate Hypothesis (H₁): Bike rentals differ significantly across weather conditions.
  - `Test`: Conduct a one-way ANOVA to evaluate whether weather conditions have a significant impact on bike rentals. Check assumptions of normality using Q-Q plots or Shapiro-Wilk test and equal variance using Levene's test.
- **Seasonal Effect on Rental Demand (One-Way ANOVA)**:

  - `Hypotheses`:
    - Null Hypothesis (H₀): The number of bike rentals is consistent across seasons.
    - Alternate Hypothesis (H₁): The number of bike rentals varies across seasons.
  - `Test`: Use one-way ANOVA to check if the season has a significant effect on the demand for electric cycles. Ensure that normality and equal variance assumptions are checked, and continue with visual analysis even if assumptions are violated.
- **Dependency Between Weather and Season (Chi-Square Test)**:

  - `Hypotheses`:
    - Null Hypothesis (H₀): Weather conditions are independent of the season.
    - Alternate Hypothesis (H₁): Weather conditions depend on the season.
  - `Test`: Perform a Chi-square test of independence using a contingency table between season and weather to determine if the two variables are related.

```python
# Function to remove outliers from numerical columns and visualize the effect with boxplots
def remove_outliers(df, numerical_columns):
    plt.figure(figsize=(15, 10))

    for i, column in enumerate(numerical_columns, 1):
        # Calculate the 5th and 95th percentiles for the column
        lower_bound = np.percentile(df[column], 5)
        upper_bound = np.percentile(df[column], 95)

        # Clip the column data to the calculated percentile bounds
        df[column] = np.clip(df[column], lower_bound, upper_bound)

        # Plot boxplot after outlier removal
        plt.subplot(3, 3, i)
        sns.boxplot(df[column])
        plt.title(f'Boxplot of {column} (After Outlier Removal)')

    plt.tight_layout()
    plt.show()

    # Return the modified DataFrame
    return df

# Apply outlier removal to the numerical columns
numerical_columns = ['datetime', 'temp', 'atemp', 'humidity', 'windspeed', 'casual', 'registered', 'count']
df = remove_outliers(df, numerical_columns)
```

# **Hypothesis Testing**

## <font color='blue'>**Check if there any significant difference between the no. of bike rides on Weekdays and Weekends?**</font>

### **Formulate Null Hypothesis ($H_0$) and Alternate Hypothesis ($H_a$)**

- **Null Hypothesis ($H_0$):** There is no significant difference in the number of bike rides between weekdays and weekends.
  - $\mu_1 = \mu_2$

- **Alternate Hypothesis ($H_a$):** There is a significant difference in the number of bike rides between weekdays and weekends.
  - $\mu_1 \neq \mu_2$

### **Select an appropriate test**

- The 2-sample independent t-test (also known as an independent t-test) is appropriate here because:
  - We are comparing the means of two independent groups (Weekdays and Weekends).
  - It checks whether the means of the two groups are statistically different.

### **Set a significance level**
- Set the significance level (α) to 0.05 (5%). This means we are 95% confident that any observed difference is statistically significant if the p-value is less than 0.05.

```python
alpha = 0.05 # 95% confidence
```

### **Calculate test Statistics / p-value**

```python
# Separate data into Weekdays and Weekends
weekdays_rides = df[df['workingday'] == 1]['count']
weekends_rides = df[df['workingday'] == 0]['count']

# Perform 2-sample t-test
t_stat, p_value = stats.ttest_ind(weekdays_rides, weekends_rides)

# Output the results
print(f'T-test statistic: {t_stat}')
print(f'P-value: {p_value}')
```

### **Decide whether to accept or reject the Null Hypothesis.**

```python
if p_value < alpha:
  print('Reject the null hypothesis (H0) and conclude that there is a significant difference in the number of bike rides between weekdays and weekends.')
else:
  print ('Fail to reject the null hypothesis, meaning there is no significant difference between the two groups.')
```

### **Draw inferences & conclusions from the analysis and provide recommendations.**

#### **Inference**
- There is no significant difference in the number of bike rides between Weekdays and Weekends.

#### **Recommendations**
- Maintain a balanced fleet and operational strategy throughout the week.
- Focus on other factors (such as weather or seasonal effects) for demand optimization instead of weekday vs. weekend differences.
- Consider uniform pricing and service promotion across all days of the week, as demand is relatively stable.

#### **Normality Test Using Shapiro-Wilk’s test**

```python
# List of unique weather conditions
weather_conditions = df['weather'].unique()

# Initialize a dictionary to store Shapiro-Wilk test results
shapiro_results = {}
alpha = 0.05  # Significance level

# Perform Shapiro-Wilk test for each weather condition
for weather in weather_conditions:
    subset = df[df['weather'] == weather]['count']

    if len(subset) >= 3:  # Shapiro-Wilk requires at least 3 data points
        stat, p_value = stats.shapiro(subset)
        shapiro_results[weather] = (stat, p_value)
    else:
        shapiro_results[weather] = (None, None)

# Print the results with interpretation
for weather, (stat, p_value) in shapiro_results.items():
    if p_value is not None:
        result = "Not Normally Distributed" if p_value <= alpha else "Normally Distributed"
        print(f'Weather Condition {weather}: Statistics = {stat:.4f}, p-value = {p_value:.4e}; {result}')
    else:
        print(f'Weather Condition {weather}: Not enough data for the test')
```

## <font color='blue'>**Check if the Weather conditions are significantly different during different Seasons?**</font>

#### **Formulate Null Hypothesis ($H_0$) and Alternate Hypothesis ($H_a$)**

- **Null Hypothesis ($H_0$):** Weather conditions are independent of the seasons. In other words, the distribution of weather conditions is the same across all seasons.

- **Alternate Hypothesis ($H_a$):** Weather conditions are not independent of the seasons. This implies that the distribution of weather conditions varies across different seasons.


### **Select an appropriate test**
- To test the association between categorical variables (Weather and Season), the `Chi-square test of independence` is appropriate.

### **Create a Contingency Table against ‘Weather’ & ‘Season’ columns**

```python
# Create a contingency table
contingency_table = pd.crosstab(df['weather'], df['season'])

# Display the contingency table
print("Contingency Table:\n", contingency_table)
```

### **Set a Significance Level and Calculate the Test Statistics / p-value**

```python
# Set significance level
alpha = 0.05

# Perform Chi-square test
chi_stat, p_value, df, exp_freq = chi2_contingency(contingency_table)

# Print results
print(f'Chi-square Test Statistic: {chi_stat}')
print(f'Chi-square Test p-value: {p_value}')
print(f'Degrees of Freedom: {df}')
print(f'Expected Frequencies:\n{exp_freq}')
```

### **Decide whether to accept or reject the Null Hypothesis.**

```python
# Decision based on significance level
if p_value <= alpha:
    print('Reject the null hypothesis: There is a significant association between weather conditions and seasons.')
else:
    print('Fail to reject the null hypothesis: There is no significant association between weather conditions and seasons.')
```

### **Draw inferences & conclusions from the analysis and provide recommendations.**


#### **Inference**

The Chi-square test results in a p-value that is significantly less than the significance level of 0.05. Therefore, we reject the null hypothesis of independence between weather conditions and seasons.
